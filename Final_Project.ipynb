```python
#Import module
from time import sleep as timeout
import pandas as pd

signment = 0
quizz = 0
jum_mahasiswa = 0
database_nilai = []
database_nama = []
database_nim = []
columns = [("PROJECT", "15%"), ("MIDTERM", "25%"), ("FINAL", "25%")]

"""Function input semua nilai yang dimiliki mahasiswa"""
def project():
  nilai_project = 0
  inp = int(input("Masukan nilai project: "))
  nilai_project = inp
  return nilai_project

def mid():
  nilai_mid = 0
  inp = int(input("Masukan nilai mid: "))
  nilai_mid = inp
  return nilai_mid

def final():
  nilai_final = 0
  inp = int(input("Masukan nilai final: "))
  nilai_final = inp
  return nilai_final

def quiz():
  while True:
    try:
      inp_quiz = int(input("Ada berapa quiz: "))
      quizz = inp_quiz
      if (inp_quiz < 5):
        break
      else:
        print("Quiz maksimal 4!")
    except ValueError:
      print("Format salah!")

  nilai_quiz = []
  for i in range(1, inp_quiz+1):
    inp = int(input(f"Nilai quiz ke {i}: "))
    nilai_quiz.append(inp)

  sord_colom= sorted([("QUIZ", f"{i+1}") for i in range(quizz)], key=lambda x: x[1])
  return nilai_quiz, sord_colom

def assignment():
  while True:
    try:
      inp_assignment = int(input("Ada berapa assignment: "))
      signment = inp_assignment
      if (inp_assignment < 5):
        break
      else:
        print("Assignment maksimal 4!")
    except ValueError:
      print("Format salah!")
  nilai_assignment = []
  for i in range(1, inp_assignment+1):
    inp = int(input(f"Nilai assignment ke {i}: "))
    nilai_assignment.append(inp)

  sord_colom= sorted([("ASSIGNMENT", f"{i+1}") for i in range(signment)], key=lambda x: x[1])

  return nilai_assignment, sord_colom

def nilai(assignment, quiz, project, mid, final):
  data_nilai = []
  data_nilai = [project, mid, final]
  data_nilai[:0] = quiz
  data_nilai[:0] = assignment
  return data_nilai

"""Menghitung Nilai"""
def hitung_nilai(database_nilai):
  assignment = sum([nilai[0] for nilai in database_nilai])
  quiz = sum([nilai[1] for nilai in database_nilai])
  project = sum([nilai[2] for nilai in database_nilai])
  mid = sum([nilai[3] for nilai in database_nilai])
  final = sum([nilai[4] for nilai in database_nilai])

  rata_assignment = assignment / len(database_nilai)
  rata_quiz = quiz / len(database_nilai)
  rata_project = project / len(database_nilai)
  rata_mid = mid / len(database_nilai)
  rata_final = final / len(database_nilai)

  grade = int((
    rata_assignment * 0.20 +
    rata_quiz * 0.15 +
    rata_project * 0.15 +
    rata_mid * 0.25 +
    rata_final * 0.25
  ))

  return grade

def huruf(grade):
  if (grade > 91):
    return 'A'
  elif (grade > 85):
    return 'A-'
  elif (grade > 82):
    return 'B+'
  elif (grade > 78):
    return 'B'
  elif (grade > 75):
    return 'B-'
  elif (grade > 70):
    return 'C+'
  elif (grade > 67):
    return 'C'
  elif (grade > 60):
    return 'C-'
  elif (grade > 40):
    return 'D'
  else:
    return 'F'


"""Proses menginput smua nilai"""
while True:
  try:
    jum_user = int(input("Masukan jumlah mahasiswa!!\n>>> "))
    jum_mahasiswa = jum_user
    for i in range(1, jum_user+1):
      while True:
        try:
          nim = int(input(f"Masukan NIM mahasiswa ke {i}: "))
          database_nim.append(nim)
          nem = str(input(f"Masukan nama mahasiswa ke {i}: "))
          database_nama.append(nem)
          nilai_quiz, sortquiz = quiz()
          nilai_assignment, sortsign = assignment()
          nilai_project = project()
          nilai_mid = mid()
          nilai_final = final()
          list_nilai = nilai(nilai_assignment, nilai_quiz, nilai_project, nilai_mid, nilai_final)
          database_nilai.append(list_nilai)
          break
        except ValueError:
          print("Format salah!")
    break
  except ValueError:
    print("Format salah!")


columns[:0] += sortquiz
columns[:0] += sortsign

#CEK ERROR!
print(columns)

#TABEL
index = pd.MultiIndex.from_tuples([(database_nim[i], database_nama[i]) for i in range(jum_mahasiswa)], names=["NIM", "NAMA"])
column = pd.MultiIndex.from_tuples(columns)
df = pd.DataFrame(database_nilai, columns=column,index=index)
df["GRADE"] = df.mean(axis=1)

print(df)
nilai_akhir = hitung_nilai(database_nilai)
huruf = huruf(nilai_akhir)
print(f"Letter Grade: {huruf}")
```
